# DecodeLabs Project 1

# Import Libraries

```python
import pandas as pd

print("Libraries Imported Successfully")
```

# Load Dataset

```python
file_path = "Dataset for Data Analytics 1.xlsx"

df = pd.read_excel(file_path)

print("\nDataset Loaded Successfully")
```

```python
print("\n========== DATASET INFO ==========")

print("\nShape of Dataset:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nFirst 5 Rows:")
print(df.head())
```

# Missing values check

```python
print("\n========== MISSING VALUES ==========")

missing_values = df.isnull().sum()

print(missing_values)

print("\nTotal Missing Values:")
print(missing_values.sum())
```

# Handling Missing Values

```python
# Fill missing CouponCode values

if "CouponCode" in df.columns:

    df["CouponCode"] = df["CouponCode"].fillna("No Coupon")

    print("\nMissing CouponCode values filled.")

else:

    print("\nCouponCode column not found.")
```

```python
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

# Duplicates Check

```python
print("\n========== DUPLICATE RECORDS ==========")

duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)
```

```python
df = df.drop_duplicates()

print("Duplicates Removed Successfully")
```

```python
if "OrderID" in df.columns:

    duplicate_ids = df["OrderID"].duplicated().sum()

    print("\nDuplicate Order IDs:", duplicate_ids)

else:

    print("\nOrderID column not found")
```

# Date Format Standardize

```python
if "Date" in df.columns:

    df["Date"] = pd.to_datetime(
        df["Date"],
        errors="coerce"
    )

    df["Date"] = df["Date"].dt.strftime("%Y-%m-%d")

    print("\nDate Format Standardized")

else:

    print("\nDate column not found")
```

# Text Cleaning

```python
text_columns = [

    "Product",
    "ShippingAddress",
    "PaymentMethod",
    "OrderStatus",
    "ReferralSource"

]

for col in text_columns:

    if col in df.columns:

        df[col] = df[col].astype(str).str.strip()


print("\nText Columns Cleaned")
```

# Numeric Formatting

```python
numeric_columns = [

    "UnitPrice",
    "TotalPrice"

]

for col in numeric_columns:

    if col in df.columns:

        df[col] = df[col].round(2)


print("\nNumeric Values Rounded")
```

# Validation

```python
print("\n========== FINAL VALIDATION ==========")

print("\nMissing Values:")
print(df.isnull().sum().sum())

if "OrderID" in df.columns:

    print(
        "Duplicate Order IDs:",
        df["OrderID"].duplicated().sum()
    )

print(
    "Duplicate Rows:",
    df.duplicated().sum()
)
```

# Save Clean Dataset

```python
output_file = "cleaned_dataset.xlsx"

df.to_excel(
    output_file,
    index=False
)

print("\nClean Dataset Saved Successfully")

print("\nFile Name:")
print(output_file)
```
